```python
#reading the data 

import pandas as pd
df = pd.read_csv("data.csv", encoding='ISO-8859-1')
df
```

```python

#Number of rows and columns the dataset has.
print("Dataset dimensions (rows, columns):", df.shape)
```

```python

#removing duplicates found in the dataset
df = df.drop_duplicates()
df
```

```python
#Check for Missing Values
missing_values = df.isnull().sum()
print("Missing values per column:")
print(missing_values)
```

```python

#Remove rows with missing values
df = df.dropna()
df
```

```python

#Finding the Negative Unit Price
negative_unit_price = df[df['UnitPrice'] < 0]
print("\nRows with negative UnitPrice:")
print(negative_unit_price)
```

```python
#Find Negative Quantity
negative_quantity = df[df['Quantity'] < 0]
print("\nRows with negative Quantity:")
print(negative_quantity)
```

```python
# Convert InvoiceDate to datetime to analyze sales by month, day, hour 
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])

```

```python
# Create TotalSales column in order to know revenue per transaction
df['TotalSales'] = df['Quantity'] * df['UnitPrice']
```

```python
# See top products by sales
top_products = df.groupby('Description')['TotalSales'].sum().sort_values(ascending=False).head(10)
print(top_products)
```

```python
# See top countries by sales
top_countries = df.groupby('Country')['TotalSales'].sum().sort_values(ascending=False).head(10)
print(top_countries)
df
```
